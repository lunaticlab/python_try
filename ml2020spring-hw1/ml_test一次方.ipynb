```python
import random
import matplotlib.pyplot as plt
yhead=[]
x=[]
for i in range(0,10):
    yhead.append(13*i+19)
    x.append(i)
print(x)
print(yhead)
plt.plot(x,yhead,'r--')
plt.show()
```

```python
y_predict=[0]*len(x)
w=0;b=0
generation=0
r=0.003 #learning rate 很重要依實測 適合小於3/1000
def differential(w,b,x,yhead,feature):
    global y_predict
    
    diff_value=0
    if feature==0:
        for i in range(0,10):
            diff_value+=2*(yhead[i]-(y_predict[i]))*(-1)
            
       
    if feature==1:
        for i in range(0,10):
            diff_value+=2*(yhead[i]-(y_predict[i]))*(-1)*(x[i])
    return diff_value       
def gradient():
    global w,b,x,yhead
    w=w-r*(differential(w,b,x,yhead,1))
    b=b-r*(differential(w,b,x,yhead,0))
    
while generation<=100:
    
    for i in range(0,10):
        y=w*x[i]+b #function
        y_predict[i]=y
    #print(y_predict)
    gradient()
    generation+=1
```

```python
plt.plot(x,yhead,'r--')
plt.plot(x,y_predict,'b--')
print(w,b)
plt.show()
```
